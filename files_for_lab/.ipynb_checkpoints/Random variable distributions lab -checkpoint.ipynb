```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
numericals = data.select_dtypes(np.number)
```

```python
numericals = pd.DataFrame(numericals)
```

```python
numericals.head()
```

```python
for column in numericals:
    sns.distplot(numericals[column])
    plt.show()
```

## conclusions:
#### customer lifetime value - exponential
#### income - normal with null values
#### monthly premium auto - exponential
#### months sisce last claim - exponential
#### months since policy inception - normal
#### number of open complains - exponential (but with a lot of null)
#### number of policies - exponential
#### total claim amount - exponential

```python
def log_transfom(x):
    if np.isfinite(x) and x>0:  
        return np.log(x)
    else:                       
        return np.NAN
```

```python
for column in numericals:  
    pd.Series(numericals[column].apply(log_transfom)).hist()
    plt.show()
```

```python
for column in numericals:  
    sns.distplot(numericals[column].apply(log_transfom))
    plt.show()
```

## conclusions:
#### they do look different...
#### customer lifetime value - looks more like normal distribution with outlieres
#### income - looks more exponential now (but to the different diarection, not sure what it means)
#### monthly premium auto - exponential
#### months sisce last claim - exponential
#### months since policy inception - exponential
#### number of open complains - exponential (but with a lot of null)
#### number of policies - exponential
#### total claim amount - looks more like normal now
